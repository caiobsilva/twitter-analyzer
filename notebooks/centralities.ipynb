```python
import networkx as nx
```

```python
graph = nx.karate_club_graph()
nx.draw(graph, node_color="#555555", edge_color="#989898")
```

```python
degree_centrality = nx.algorithms.degree_centrality(graph)
adjusted_centrality = { k: v*1000 for k, v in degree_centrality.items() }
print(adjusted_centrality)
nx.draw(graph, node_size=list(adjusted_centrality.values()), node_color="#555555", edge_color="#989898")
```

```python
betweenness_centrality = nx.algorithms.betweenness_centrality(graph)
adjusted_centrality = { k: v*1000 if v*1000 > 5.0 else 5.0 for k, v in betweenness_centrality.items() }
print(adjusted_centrality)
nx.draw(graph, node_size=list(adjusted_centrality.values()), node_color="#555555", edge_color="#989898")
```

```python
closeness_centrality = nx.algorithms.closeness_centrality(graph)
adjusted_centrality = { k: v**2*1000 for k, v in closeness_centrality.items() }
print(adjusted_centrality)
nx.draw(graph, node_size=list(adjusted_centrality.values()), node_color="#555555", edge_color="#989898")
```

```python
eigenvector_centrality = nx.algorithms.eigenvector_centrality(graph)
adjusted_centrality = { k: v*1000 for k, v in eigenvector_centrality.items() }
print(adjusted_centrality)
nx.draw(graph, node_size=list(adjusted_centrality.values()), node_color="#555555", edge_color="#989898")
```
